```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import sqrt
import time
```

```python
def AddCircle(x, y, r, color):
    c = plt.Circle((x, y), radius = r, color=color, linewidth=2, fill=False)
    ax = plt.gca()
    ax.add_patch(c)
```

```python
circles = []

with open('in.txt', 'r') as infile:
    for i, line in enumerate(infile):
        a = map(int, line.split())
        if i == 0:
            x1, y1, x2, y2 = a
            mx = (x1 + x2) / 2.0
            my = (y1 + y2) / 2.0
            a = - (x2 - x1)
            b = (y2 - y1)
            c = sqrt(a ** 2 + b ** 2)
            sn = a / c
            cs = b / c
            rMin = c / 2
            rx = rMin
            ry = 1E12
        if i >= 2:
            x, y, r = a
            circles.append((x, y, r))
            
print mx, my, sn, cs, rx, ry
print circles
```

```python
def CheckCircle(cx, cy, R):
    for c in circles:
        x, y, r = c
        d = sqrt((x - cx) ** 2 + (y - cy) ** 2)
        if d > R - r and d < R + r:
            return 0
    return 1
```

```python
def Check(R):
    l = sqrt(R ** 2 - rMin ** 2)
    
    plt.clf()
    
    for c in circles:
        x, y, r = c
        AddCircle(x, y, r, 'blue')
    
    AddCircle(mx + l * cs, my + l * sn, R, 'red')
    AddCircle(mx - l * cs, my - l * sn, R, 'red')
    
    plt.axis('scaled')
    plt.show()
    
    return CheckCircle(mx + l * cs, my + l * sn, R) + CheckCircle(mx - l * cs, my - l * sn, R)
```

```python
for _ in range(200):
    rm = (rx + ry) / 2

    cnt = Check(rm)
    print rm, cnt

    if cnt > 0:
        ry = rm
    else:
        rx = rm
```
