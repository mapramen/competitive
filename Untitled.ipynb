```python
%matplotlib inline
import matplotlib.pyplot as plt
from math import sqrt
import time
import numpy as np
```

```python
def AddCircle(x, y, r, color):
    c = plt.Circle((x, y), radius = r, color=color, linewidth=2, fill=False)
    ax = plt.gca()
    ax.add_patch(c)
```

```python
circles = []

x1, x2, y1, y2 = 1000, -1000, 1000, -1000
with open('in.txt', 'r') as infile:
    for i, line in enumerate(infile):
        a = map(int, line.split())
        if i > 0:
            x, y, r = a
            x1 = min(x1, x - r)
            x2 = max(x2, x + r)
            y1 = min(y1, y - r)
            y2 = max(y2, y + r)
            circles.append((x, y, r))

print (x1, x2), (y1, y2)
print circles
```

```python
plt.clf()

num_points = 1000
points = np.random.uniform((x1, y1), (x2, y2), (num_points, 2))

masks = set([])

for circle in circles:
    x, y, r = circle
    AddCircle(x, y, r, 'green')
    
for point in points:
    x, y = point
    mask = 0
    for i, circle in enumerate(circles):
        cx, cy, cr = circle
        if (cx - x) ** 2 + (cy - y) ** 2 < cr ** 2:
            mask |= (1 << i)
        if mask != 0:
            plt.scatter(x, y, c = 'red', marker='.')
    masks.add(mask)

plt.axis('scaled')
plt.show()

print masks
print len(masks)
```
